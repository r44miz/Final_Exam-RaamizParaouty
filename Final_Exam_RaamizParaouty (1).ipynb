```python
# FINAL EXAM - Task Manager
# Compliant with Weeks 3â€“6 techniques: loops, lists, input, basic error handling, doctest

tasks = []  # Week 4 - list for task storage

def display_menu():
    """
    Display menu options - Week 3 formatting

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")

def get_user_choice():
    """
    Get and validate user choice - Week 6 input validation

    >>> get_user_choice()  # doctest: +SKIP
    """
    while True:
        choice = input("Choose an option (1-4): ").strip()
        if choice in ["1", "2", "3", "4"]:
            return choice
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

def add_task():
    """
    Add a task to the global tasks list

    >>> tasks.clear()
    >>> add_task()  # doctest: +SKIP
    Enter task: Buy groceries
    """
    task = input("Enter task: ").strip()
    if task:
        tasks.append(task)
        print(f"Added: {task}")
    else:
        print("Task cannot be empty.")

def list_tasks():
    """
    List all tasks

    >>> tasks[:] = ['Task A', 'Task B']
    >>> list_tasks()  # doctest: +SKIP
    1. Task A
    2. Task B
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")

def remove_task():
    """
    Remove a task by its number

    >>> tasks[:] = ['Sample Task']
    >>> remove_task()  # doctest: +SKIP
    Enter task number to remove: 1
    """
    if not tasks:
        print("No tasks to remove.")
        return

    try:
        index = int(input("Enter task number to remove: ")) - 1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            print(f"Removed: {removed}")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

def main():
    """Main programme loop - Week 5 loop patterns only"""
    print("Welcome to the Task Manager!")
    while True:
        display_menu()
        choice = get_user_choice()

        if choice == "1":
            add_task()
        elif choice == "2":
            list_tasks()
        elif choice == "3":
            remove_task()
        elif choice == "4":
            print("Goodbye!")
            break

if __name__ == "__main__":
    import doctest
    doctest.testmod(verbose=True)
    main()
```
